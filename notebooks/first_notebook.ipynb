```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt


sp.init_printing()

# Symbolic variable
x = sp.symbols('x')

# Function and derivative
f = x**3
df = sp.diff(f, x)
ddf = sp.diff(df, x)

print (sp.latex(f))
print (sp.latex(df))
print (sp.latex(ddf))
expr = sp.Integral(sp.sin(x) / x, x)
expr


f_num = sp.lambdify(x, f, 'numpy')
df_num = sp.lambdify(x, df, 'numpy')
ddf_num = sp.lambdify(x, ddf, 'numpy')

# Define x values
x_vals = np.linspace(-5, 5, 500)

# Evaluate functions
y_vals = f_num(x_vals)
dy_vals = df_num(x_vals)
ddy_vals = ddf_num(x_vals)

# Plot
plt.plot(x_vals, y_vals, label='f(x) = x^3')
plt.plot(x_vals, dy_vals, label="f'(x) = 3x^2")
plt.plot(x_vals, ddy_vals, label="f''(x) = 6x")
    
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and its derivative')
plt.grid(True)
plt.legend()
plt.show()
```
